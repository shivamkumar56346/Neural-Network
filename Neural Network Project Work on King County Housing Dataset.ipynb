```python
df=df.drop(['date'],axis=1)
df.head(5)
```

```python
df=df.drop(['zipcode'],axis=1)
df.head(5)
```

```python
#Features
x=df.drop(['price'],axis=1)
x.head()
```

```python
# Converted Feature Dataframe to Numpy
x=x.values
print(x)
```

```python
# Label
y=df['price']
y.head()
```

```python
# Label DataFrame to Numpy
y=y.values
print(y)
```

```python
# Split our dataset into train & test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# Data Normalization with MinMaxScaler to prevent data leak.
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Importing Libraries for creating a neural network
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout,BatchNormalization
```

```python
# Creating a neural network layers
model=Sequential()
model.add(Dense(20,input_shape=(20,),activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(20,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1))
```

```python
from keras import optimizers
adam=optimizers.Adam(lr=0.01)
model.compile(optimizer=adam,loss='mean_squared_error',metrics=['mse'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=200,verbose=1)
```

```python
model.history.history
```

```python
loss_df=pd.DataFrame(model.history.history)
loss_df.head()
```

```python
plt.figure(figsize=(15,8))
loss_df.plot()
```

```python
model.evaluate(x_test,y_test)
```

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
print("MAE : ",mean_absolute_error(y_test,predictions))
print("MSE : ",mean_squared_error(y_test,predictions))
print("RMSE : ",np.sqrt(mean_squared_error(y_test,predictions)))
print("Explained Variance : ",explained_variance_score(y_test,predictions))
```

```python
# PART TWO 
# Creating a new ann model by building an another Neural Network.
```

```python
def my_ann_model():
    model=Sequential()
    model.add(Dense(20,input_shape=(20,),activation='relu'))
    model.add(Dense(20,activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(20,activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(20,activation='relu'))
    model.add(BatchNormalization())
    model.add(Dense(1))
    adam=optimizers.Adam(lr=0.01)
    model.compile(optimizer=adam,loss='mean_squared_error',metrics=['mse'])
    return model
```

```python
new_model=my_ann_model()
```

```python
new_model.summary()
```

```python
# Using Callbacks & Early Stopping for improving our model performance on test data.
from keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
new_model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=400,verbose=1,callbacks=[early_stop])
```

```python
new_model.history.history
```

```python
new_loss_df=pd.DataFrame(new_model.history.history)
new_loss_df.head()
```

```python
plt.figure(figsize=(15,8))
new_loss_df.plot()
```

```python
new_model.evaluate(x_test,y_test)
```

```python
prediction=new_model.predict(x_test)
prediction
```

```python
print("MAE : ",mean_absolute_error(y_test,prediction))
print("MSE : ",mean_squared_error(y_test,prediction))
print("RMSE : ",np.sqrt(mean_squared_error(y_test,prediction)))
print("Explained Variance : ",explained_variance_score(y_test,prediction))
```

```python
# Part Three
# Using Keras Regressor & GridSearchCV for HyperParameter Optimization & Tuning.
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
third_model=KerasRegressor(build_fn=my_ann_model,verbose=1)
```

```python
batch_size=[64,128]
epochs=[200,400]
param_grid=dict(batch_size=batch_size,epochs=epochs)
print(param_grid)
```

```python
grid=GridSearchCV(estimator=third_model,param_grid=param_grid,cv=2,scoring='r2',verbose=1,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
print("Best: %f using %s" % (grid.best_score_,grid.best_params_))
```

```python
grid_predictions=grid.predict(x_test)
```

```python
print(grid_predictions)
```

```python
print("MAE : ",mean_absolute_error(y_test,grid_predictions))
print("MSE : ",mean_squared_error(y_test,grid_predictions))
print("RMSE : ",np.sqrt(mean_squared_error(y_test,grid_predictions)))
print("Explained Variance : ",explained_variance_score(y_test,grid_predictions))
```
